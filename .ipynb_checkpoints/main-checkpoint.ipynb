```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
df = pd.read_csv("marks.csv")
```

```python
df.head()
```

```python
df = df.set_axis(["Name", "roll number","Department","Email","Mid 1","Quiz 1","Quiz 2","Mid 2","Quiz 3"], axis=1)
df.head(1)
```

```python
df.replace(['-'],0, inplace = True)
```

```python
#converting the data type to int
df['Quiz 1'] = pd.to_numeric(df['Quiz 1'], errors="coerce")
df['Quiz 2'] = pd.to_numeric(df['Quiz 2'], errors="coerce")
df['Quiz 3'] = pd.to_numeric(df['Quiz 3'], errors="coerce")
```

```python
#converting the data tyope to int
df['Mid 1'] = pd.to_numeric(df['Mid 1'], errors="coerce")
df['Mid 2'] = pd.to_numeric(df['Mid 2'], errors="coerce")
```

```python
#calculating the total quiz marks
df['Total quizzes'] = df['Quiz 1'] + df['Quiz 2'] + df['Quiz 3']
```

```python
#calculating the total mid marks
df['Total mids'] = df['Mid 1'] + df['Mid 2']
```

```python
df. drop("Total quiz", axis=1, inplace=True)
```

```python
df.head()
```

```python
#calculating the total marks
df["Total internal"] = df["Total mids"]+ df["Total quizzes"]
```

```python
final_data = df.filter(["Email address", "roll number", "Name", "Total mids", "Total quizzes", "Total internal"], axis=1)
```

```python
final_data['roll number'] = final_data['roll number'].str.upper()
```

```python
final_data.set_index("roll number", inplace=True)
```

```python
final_data.head()
```

```python
#mapping the grades
grades = {
    45: "O",
    40: "A+",
    35: "A",
    30: "B+",
    25: "B",
    20: "C",
    15: "P",
    0: "F",  
}


def grade_mapping(value):
    """Map numerical grade to letter grade."""
    for key, letter in grades.items():
        if value >= key:
            return letter


letter_grades = final_data["Total internal"].map(grade_mapping)
final_data["Final Grade"] = pd.Categorical(
    letter_grades, categories=grades.values(), ordered=True
)
```

```python
final_data.head()
```

```python
final_data[90:103]
```

```python
#plotting th graph of the grades for all sections
grade_counts = final_data["Final Grade"].value_counts().sort_index()
grade_counts.plot.bar()
plt.show()
```

```python
#*****
final_data["Total internal"].plot.hist(bins=20, label="Histogram")
final_data["Total internal"].plot.density(
    linewidth=4, label="Kernel Density Estimate"
)

final_mean = final_data["Total internal"].mean()
final_std = final_data["Total internal"].std()
x = np.linspace(final_mean - 5 * final_std, final_mean + 5 * final_std, 200)
normal_dist = scipy.stats.norm.pdf(x, loc=final_mean, scale=final_std)
plt.plot(x, normal_dist, label="Normal Distribution", linewidth=4)
plt.legend()
plt.show()
```

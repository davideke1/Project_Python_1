```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
df = pd.read_csv("marks.csv")
```

```python
grades = {
    35: "A",
    30: "B",
    25: "C",
    23: "D",
    0: "F",
}


def grade_mapping(value):
    """Map numerical grade to letter grade."""
    for key, letter in grades.items():
        if value >= key:
            return letter


letter_grades = df["total_internal"].map(grade_mapping)
df["Final Grade"] = pd.Categorical(
    letter_grades, categories=grades.values(), ordered=True
)
```

```python

df["total_internal"].plot.hist(bins=20, label="Histogram")
df["total_internal"].plot.density(
    linewidth=4, label="Kernel Density Estimate"
)

final_mean = df["total_internal"].mean()
final_std = df["total_internal"].std()
x = np.linspace(final_mean - 5 * final_std, final_mean + 5 * final_std, 200)
normal_dist = scipy.stats.norm.pdf(x, loc=final_mean, scale=final_std)
plt.plot(x, normal_dist, label="Normal Distribution", linewidth=4)
plt.legend()
plt.show()
```

```python
quiz_scores = df.filter(regex=r"^Module \d$", axis=1)
quiz_max_points = pd.Series(
    {"Module 1": 10, "Module 2": 10, "Module 3": 10}
)
sum_of_quiz_scores =quiz_scores.sum(axis=1)
sum_of_quiz_max = quiz_max_points.sum()
final_data["Total Modules"] = sum_of_quiz_scores / sum_of_quiz_max

average_quiz_scores = (quiz_scores / quiz_max_points).sum(axis=1)
final_data["Average Modules"] = average_quiz_scores / quiz_scores.shape[1]

final_data["Module Score"] = final_data[
    ["Total Modules", "Average Modules"]
].max(axis=1)
```
